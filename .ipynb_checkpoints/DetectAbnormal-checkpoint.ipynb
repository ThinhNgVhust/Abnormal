```python
class DetectAbnormal:
    zip_file = "__csv_ch_a_l.zip"
    normal_file = "normal"
    abnormal_file = "abnormal"
    num_of_field= 12
    def __init__(self):
        self.zip_file = DetectAbnormal.zip_file
        self.lst_normal = []
        self.lst_abnormal = []
        self.unzip()
        self.X_normal = self.get_data(DetectAbnormal.normal_file,self.lst_normal)
        self.X_abnormal = self.get_data(DetectAbnormal.abnormal_file,self.lst_abnormal)
    
    def unzip(self):
        import zipfile
        with zipfile.ZipFile(self.zip_file, 'r') as zip_ref:
            zip_ref.extractall(os.getcwd())
    
    def get_data(self,normal_file,lst_file): 
        list_csv_file = os.path.join(os.getcwd(),normal_file)
        X = np.random.randint(10, size=(1, 12))
        print("Getting data from {} folder".format(normal_file))
        if os.path.isdir(list_csv_file):
            for file_name in os.listdir(list_csv_file):
                lst_file.append(file_name)
                dataFrame = pd.read_csv(os.path.join(list_csv_file,file_name))
                numpy_arr = dataFrame.to_numpy()[:,:12]
    #             numpy_arr = numpy_arr[25:90,:]
                X = np.concatenate([X,numpy_arr])
        print("Done!")
        return X.copy()[1:,:]  
```

# Features Importance using Sklearn

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
forest = ExtraTreesClassifier(n_estimators=250,
                              random_state=0)
```

```python
forest.fit(X, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")
```

```python
for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Plot the impurity-based feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
        color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```

```python
polynomial_svm_clf = Pipeline([
("poly_features", PolynomialFeatures(degree=4)),
("scaler", StandardScaler()),
("svm_clf", LinearSVC(C=10, loss="hinge"))
])
polynomial_svm_clf.fit(X_train, y_train)
```

```python
y_pre = polynomial_svm_clf.predict(X_test)

print("Precision: {}".format(precision_score(y_test,y_pre)))
print("Recall: {}".format(recall_score(y_test,y_pre)))
print("Scores: {}".format(polynomial_svm_clf.score(X_test,y_test)))
```

```python
from sklearn.metrics import confusion_matrix

y_pre = rbf_kernel_svm_clf.predict(X_test)
print("Scores: {}".format(rbf_kernel_svm_clf.score(X_test,y_test)))
print("Precision: {}".format(precision_score(y_pre,y_test)))
print("Recall: {}".format(recall_score(y_test,y_pre)))
print("Confusion: ")
confusion_matrix(y_test,y_pre)
```

## CNN

```python
import tensorflow
from tensorflow import keras
from keras.utils import to_categorical
```

```python
print(tensorflow.__version__)
print(keras.__version__)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
print("X_train.shape: {}".format(X_train.shape))
print("X_test.shape: {}".format(X_test.shape))
y_train= y_train.astype(int)
y_test= y_test.astype(int)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
print(y_train)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(13,input_shape=[1, 12]))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(16, activation="relu"))
model.add(keras.layers.Dense(16, activation="relu"))
model.add(keras.layers.Dense(32, activation="relu"))
model.add(keras.layers.Dense(32, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(128, activation="relu"))
model.add(keras.layers.Dense(256, activation="relu"))
model.add(keras.layers.Dense(512, activation="relu"))
model.add(keras.layers.Dense(1024, activation="relu"))
model.add(keras.layers.Dense(512, activation="relu"))
model.add(keras.layers.Dense(128, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(32, activation="relu"))
model.add(keras.layers.Dense(32, activation="relu"))
model.add(keras.layers.Dense(16, activation="relu"))
model.add(keras.layers.Dense(16, activation="relu"))
model.add(keras.layers.Dense(12, activation="relu"))
model.add(keras.layers.Dense(2, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',
loss='categorical_crossentropy',
metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=30,validation_data=(X_test, y_test))
```
